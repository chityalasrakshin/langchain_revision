```python
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("why is the best movie in tollywood")
response.content
```

Stream

```python
model.stream("Write me a 200words paragraph on Data Engineering")
```

```python
for chunk in model.stream("Write me a 200words paragraph on Data Engineering"):
    print(chunk.text, end="", flush=True)
```

BATCH

```python
response = model.batch([
    "what is current trending software tech",
    "what is current trending skills in sw industry",
    "High paying jobs till date"
])
```
